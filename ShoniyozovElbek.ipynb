```python
from random import choice
import numpy as np
fig = plt.figure(figsize=(20,15))
asian_countries = [
    "Afghanistan", "Armenia", "Azerbaijan", "Bahrain", "Bangladesh",
    "Bhutan", "Brunei", "Cambodia", "China", "Cyprus", "Georgia",
    "India", "Indonesia", "Iran", "Iraq", "Israel", "Japan", "Jordan",
    "Kazakhstan", "Kuwait", "Kyrgyzstan", "Laos", "Lebanon", "Malaysia",
    "Maldives", "Mongolia", "Myanmar", "Nepal", "North Korea",
    "Oman", "Pakistan", "Palestine", "Philippines", "Qatar",
    "Saudi Arabia", "Singapore", "South Korea", "Sri Lanka", "Syria",
    "Taiwan", "Tajikistan", "Thailand", "Timor-Leste", "Turkey",
    "Turkmenistan", "United Arab Emirates", "Uzbekistan", "Vietnam", "Yemen"
]
colors_rgb = np.array([
    (0, 0, 0),
    (255, 255, 255),
    (255, 0, 0),
    (0, 255, 0),
    (0, 0, 255),
    (255, 255, 0),
     (0, 255, 255),
     (255, 0, 255),
     (255, 165, 0),
     (255, 192, 203),
     (128, 0, 128),
     (165, 42, 42),
     (144, 238, 144),
     (135, 206, 235),
     (230, 230, 250),
     (250, 128, 114),
     (64, 224, 208),
     (255, 215, 0),
     (0, 128, 128),
     (128, 128, 0),
     (128, 0, 0),
     (0, 0, 128),
     (75, 0, 130),
     (112, 128, 144),
     (160, 82, 45),
     (255, 99, 71),
     (70, 130, 180),
     (153, 50, 204)
])/255
def x_y(df, country):
    df_c = df[df["country"] ==  country]  
    y =  df_c["Median Age"]
    x = df_c["Year"]
    return x, y
    
for country in asian_countries:    
    x, y = x_y(df, country)     
    plt.plot(x,y, c =  choice(colors_rgb)  )
    
   
plt.legend(asian_countries)    
plt.show()
```
